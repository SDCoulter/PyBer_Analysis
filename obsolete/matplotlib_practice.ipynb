```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create our plot.
plt.plot(x_axis, y_axis, label="Boston")
# Create labels for our axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set a limit of 0-45.
plt.ylim(0, 45)
# Add a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# And a look at the OO Interface.
fig = plt.figure()
ax = fig.add_subplot()
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_title("PyBer Fare by Month")
ax.grid()
ax.set_ylim(0, 45)
ax.plot(x_axis, y_axis, marker="D", linewidth=2, color="green", label="Boston")
ax.legend()
```

```python
# Another quick way of producing this graph:
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create our plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")
# Create labels for our axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set a limit of 0-45.
plt.ylim(0, 45)
# Add a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Re-iterating the x and y values.

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a bar chart.
plt.bar(x_axis, y_axis)
```

```python
plt.bar(x_axis, y_axis, color="green", label="Boston")
# Create labels.
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Title
plt.title("PyBer Fare by Month")
# And a legend.
plt.legend()
```

```python
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
# Create labels.
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Title
plt.title("PyBer Fare by Month")
# And a legend.
plt.legend()

# Invert the y-values.
plt.gca().invert_yaxis()
```

```python
# OO method.
# Create the plot:
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# And a horizontal version:
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="c", label="Chicago")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
```

```python
# MATLAB version of Scatter Graphs.
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.scatter(y_axis, x_axis, c="r", label="Chicago")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
# And using the OOI method.
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c="skyblue", s=[i * 10 for i in y_axis], alpha=0.8, edgecolors='black', linewidth=2, label="Boston")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.set_xlim(0, 50)
ax.invert_yaxis()
```
